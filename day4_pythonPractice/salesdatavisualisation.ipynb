```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = np.array([1300,1500,1600,2000,1900,1800,2100,1750,2100,1900])
days = np.arange(1,11)
```

```python
print(sales)
```

```python
total_sales = np.sum(sales)
avg_sales = np.mean(sales)
highest_sales = np.max(sales)
lowest_sales = np.min(sales)

print(total_sales)
print(avg_sales)
print(highest_sales)
print(lowest_sales)
```

```python
plt.figure(figsize=(5, 3))

plt.plot(days, sales, marker = 'o', linestyle = '-', color = 'b', label = 'Sales')

plt.axhline(avg_sales, color='r', linestyle = '--', label = 'Average Sales')
 
plt.xlabel("Days")

plt.ylabel("Sales Revenue ($)")

plt.title("Sales Trend Over 10 Days")

plt.legend()

plt.show()
 
```

```python
# Bar Chart - sales comparison by day
 
plt.figure(figsize=(5, 3))

plt.bar(days, sales, color = 'b', label = 'Sales')
 
plt.xlabel("Days")

plt.ylabel("Sales Revenue ($)")

plt.title("Sales per Day")

plt.legend()

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
 
```
