```python
#activity work
df =  pd.read_csv('STUDENT_DATA.csv')
print(df.head())
```

```python
# for fill the empty data we use 
# df.fillna(syntax)
```

```python
df.isna()
```

```python
df['total'] = df[['math','science','english']].sum(axis=1) #sum of math, science and english

df.head()
```

```python
def assign_grade(total):
    if total >= 250:
        return 'A'
    elif total >= 200:
        return 'B'
    elif total >= 150:
        return 'C'
    else:
        return 'D'
        
assign_grade(100)
df['grade'] = df['total'].apply(assign_grade)

print(df.head())
```

```python
# filter the student with grade A
grade_A_student = df[df['grade'] == 'B']
print(grade_A_student)
```

```python
print(f"how many student have geade B : {len(grade_A_student)}")
```

```python
grade_A_student = df[df['grade'] == 'D']
print(f"how many student are failed : {len(grade_A_student)}")
```

```python
math_max = max(df['math'])
print(math_max)
```

```python
print(df[df['math'] == math_max])
```
